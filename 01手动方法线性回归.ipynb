```python
import pandas as pd

url = "https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv"
data = pd.read_csv(url)
# y = w*x + b
x = data["YearsExperience"]
y = data["Salary"]

data
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y,marker = "x",color = "red")
plt.title("yearExperience-salary")
plt.xlabel("year experience")
plt.ylabel("salary (k)")
```

```python
w=0
b=0
def plot_pred(w,b):
  y_pred = x*w+b
  plt.plot(x,y_pred,color ="blue",label ="predict")
  plt.scatter(x,y,color ="red",label ="real")
  plt.xlabel("year experience")
  plt.ylabel("salary (k)")
  plt.legend()
  plt.show()

plot_pred(0,0)
```

```python
from ipywidgets import interact
interact(plot_pred,w=(-100,100,1),b=(-100,100,1))
```

```python
w=10
b=0
y_pred = w*x+b
cost = (y-y_pred)**2
cost.sum()/len(x)
```

```python
def compute_cost(x,y,w,b):
  y_pred = w*x+b
  cost = (y-y_pred)**2
  cost=cost.sum()/len(x)

  return cost




```

```python
compute_cost(x,y,10,0)
```

```python
costs = []
for w in range(-100,101):
  cost = compute_cost(x,y,w,0)
  costs.append(cost)
costs
```

```python
import matplotlib.pyplot as plt
plt.plot(range(-100,101),costs)
plt.title("cost fn for b=0,w=-100,100")
plt.xlabel("w")
plt.ylabel("cost")
plt.show()
```

```python
import numpy as np
ws=np.arange(-100,101)
bs=np.arange(-100,101)

costs = np.zeros((201,201))
i =0
for w in ws:
  j=0
  for b in bs:
    cost = compute_cost(x,y,w,b)
    costs[i,j] = cost
    j=j+1
  i+=1
costs


```

```python
ax = plt.axes(projection = "3d")
# ax.view_init(0,45)
# ax.xaxis.set_pane_color((0,0,0))
# ax.yaxis.set_pane_color((0,0,0))
# ax.zaxis.set_pane_color((0,0,0))

b_grid,w_grid = np.meshgrid(bs,ws)
ax.plot_surface(w_grid,b_grid,costs,cmap="Spectral_r",alpha =0.7)
ax.set_title("w b corresponded cost")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("cost")

w_index, b_index = np.where(costs ==np.min(costs))
plt.show()
print (ws[w_index],bs[b_index])
```
